```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Cargar los datos desde el archivo Excel
df = pd.read_excel('VentasHistoricas-CursoForecast-nuevo.xlsx', sheet_name='FORECAST-Ventas', usecols='A:B')

# Asumiendo que la columna A es el tiempo (X) y la columna B es el valor (Y)
X = np.arange(len(df)).reshape(-1, 1)  # Usar índice numérico para X
Y = df.iloc[:, 1].values  # Valores de Y

# Calcular la regresión lineal
model = LinearRegression()
model.fit(X, Y)

# Obtener los coeficientes beta_0 y beta_1
beta_0 = model.intercept_
beta_1 = model.coeficiente_[0]

# Imprimir la ecuación de la regresión
print(f"Ecuación de la regresión lineal: Y = {beta_0:.2f} + {beta_1:.2f} * X")

# Graficar los datos y la línea de regresión
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
sns.regplot(x=np.arange(len(df)), y=Y, ci=None, color='darkblue', line_kws={'color': 'red'})

# Personalizar el gráfico
plt.title('Tendencia de Ventas', fontsize=16)
plt.xlabel('Tiempo (meses)', fontsize=14)
plt.ylabel('Ventas', fontsize=14)
plt.xticks(np.arange(len(df)), df.iloc[:, 0], rotation=45)
plt.tight_layout()

# Mostrar el gráfico
plt.show()
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
from sklearn.linear_model import LinearRegression
import requests
from io import BytesIO
import warnings
warnings.filterwarnings('ignore')

# Descargar el archivo
url = "https://github.com/santiagonajera/MODELACION-Y-PRONOSTICOS-DE-LA-DEMANDA/raw/refs/heads/main/VentasHistoricas-CursoForecast-nuevo.xlsx"
response = requests.get(url)
data = BytesIO(response.content)
df = pd.read_excel(data, sheet_name='FORECAST-Ventas', usecols='A:B')
df.columns = ['Mes', 'Ventas']

print("=" * 60)
print("ANÁLISIS DE DATOS HISTÓRICOS")
print("=" * 60)
print(f"Período de datos: {df['Mes'].iloc[0]} hasta {df['Mes'].iloc[-1]}")
print(f"Total de observaciones: {len(df)}")
print(f"Ventas promedio: ${df['Ventas'].mean():,.2f}")
print("=" * 60)

# Descomposición multiplicativa
decomposition = seasonal_decompose(df['Ventas'], model='multiplicative', period=12)

# Calcular el forecast histórico
forecast_historico = decomposition.trend * decomposition.seasonal

# PROYECCIÓN HACIA EL FUTURO (12 meses)
# 1. Extraer factores estacionales promedio
seasonal_pattern = decomposition.seasonal.dropna()
# Tomar los últimos 12 valores de estacionalidad (un ciclo completo)
seasonal_factors = seasonal_pattern[-12:].values

# 2. Proyectar la tendencia usando regresión lineal
trend_data = decomposition.trend.dropna()
X = np.arange(len(trend_data)).reshape(-1, 1)
y = trend_data.values

# Entrenar modelo de regresión
model = LinearRegression()
model.fit(X, y)

# Proyectar tendencia para los próximos 12 meses
future_X = np.arange(len(trend_data), len(trend_data) + 12).reshape(-1, 1)
future_trend = model.predict(future_X)

# 3. Aplicar estacionalidad al forecast futuro
forecast_futuro = future_trend * seasonal_factors

# Crear DataFrame con resultados futuros
meses = ['Ene', 'Feb', 'Mar', 'Abr', 'May', 'Jun',
         'Jul', 'Ago', 'Sep', 'Oct', 'Nov', 'Dic']
# Determinar el año siguiente
ultimo_mes = df['Mes'].iloc[-1]
if isinstance(ultimo_mes, str):
    # Extraer año si viene en formato texto
    try:
        anio_actual = int(ultimo_mes.split('-')[0])
    except:
        anio_actual = 2024
else:
    anio_actual = ultimo_mes.year

anio_siguiente = anio_actual + 1

meses_futuros = [f"{mes}-{anio_siguiente}" for mes in meses]

df_futuro = pd.DataFrame({
    'Mes': meses_futuros,
    'Forecast': forecast_futuro
})

print("\nFORECAST PARA LOS PRÓXIMOS 12 MESES")
print("=" * 60)
print(df_futuro.to_string(index=False))
print("=" * 60)
print(f"Total proyectado: ${forecast_futuro.sum():,.2f}")
print(f"Promedio mensual proyectado: ${forecast_futuro.mean():,.2f}")
print("=" * 60)

# ============================================================
# GRÁFICAS
# ============================================================

sns.set(style="whitegrid", palette="deep")

# GRÁFICA 1: Serie Real vs Forecast Histórico
plt.figure(figsize=(16, 6))
plt.subplot(1, 2, 1)
plt.plot(df['Ventas'], label='Serie Real', color='#2E86AB', linewidth=2.5, marker='o', markersize=4)
plt.plot(forecast_historico, label='Forecast Histórico', color='#A23B72',
         linestyle='--', linewidth=2.5, marker='s', markersize=4)
plt.title('Serie Real vs. Forecast Histórico', fontsize=14, fontweight='bold')
plt.xlabel('Tiempo (meses)', fontsize=12)
plt.ylabel('Ventas', fontsize=12)
plt.legend(fontsize=10)
plt.grid(True, alpha=0.3)

# GRÁFICA 2: Componentes de la Descomposición
plt.subplot(1, 2, 2)
plt.plot(decomposition.trend.dropna(), label='Tendencia', color='#F18F01', linewidth=2.5)
plt.plot(decomposition.seasonal, label='Estacionalidad', color='#6A994E', linewidth=2, alpha=0.7)
plt.title('Componentes: Tendencia y Estacionalidad', fontsize=14, fontweight='bold')
plt.xlabel('Tiempo (meses)', fontsize=12)
plt.ylabel('Valor', fontsize=12)
plt.legend(fontsize=10)
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# GRÁFICA 3: Proyección Futura (12 meses)
plt.figure(figsize=(16, 7))

# Concatenar histórico y futuro para mejor visualización
n_historico = len(df)
n_futuro = len(df_futuro)

# Graficar datos históricos
plt.plot(range(n_historico), df['Ventas'],
         label='Ventas Históricas', color='#2E86AB', linewidth=2.5, marker='o', markersize=5)

# Graficar forecast histórico
plt.plot(range(n_historico), forecast_historico,
         label='Forecast Histórico', color='#A23B72',
         linestyle='--', linewidth=2, alpha=0.7)

# Graficar forecast futuro
future_indices = range(n_historico, n_historico + n_futuro)
plt.plot(future_indices, forecast_futuro,
         label='Forecast Futuro (12 meses)', color='#F18F01',
         linewidth=3, marker='D', markersize=7)

# Sombrear área futura
plt.axvspan(n_historico-0.5, n_historico + n_futuro, alpha=0.15, color='orange',
            label='Período Proyectado')

# Línea vertical de separación
plt.axvline(x=n_historico-0.5, color='red', linestyle=':', linewidth=2, alpha=0.7)

plt.title('Forecast Completo: Histórico + Proyección 12 Meses',
          fontsize=16, fontweight='bold')
plt.xlabel('Tiempo (meses)', fontsize=13)
plt.ylabel('Ventas', fontsize=13)
plt.legend(fontsize=11, loc='best')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

# GRÁFICA 4: Comparación de Estadísticas
plt.figure(figsize=(12, 5))

categorias = ['Histórico', 'Proyección']
promedios = [df['Ventas'].mean(), forecast_futuro.mean()]
maximos = [df['Ventas'].max(), forecast_futuro.max()]
minimos = [df['Ventas'].min(), forecast_futuro.min()]

x = np.arange(len(categorias))
width = 0.25

plt.bar(x - width, promedios, width, label='Promedio', color='#2E86AB')
plt.bar(x, maximos, width, label='Máximo', color='#F18F01')
plt.bar(x + width, minimos, width, label='Mínimo', color='#6A994E')

plt.xlabel('Período', fontsize=13)
plt.ylabel('Ventas', fontsize=13)
plt.title('Comparación Estadística: Histórico vs Proyección', fontsize=14, fontweight='bold')
plt.xticks(x, categorias)
plt.legend()
plt.grid(True, axis='y', alpha=0.3)
plt.tight_layout()
plt.show()

# GRÁFICA 5: Patrón Estacional
plt.figure(figsize=(12, 5))
plt.plot(meses, seasonal_factors, marker='o', markersize=10,
         linewidth=3, color='#6A994E', linestyle='-')
plt.axhline(y=1.0, color='red', linestyle='--', linewidth=2, alpha=0.7, label='Línea Base (1.0)')
plt.fill_between(range(12), seasonal_factors, 1, alpha=0.3, color='#6A994E')
plt.title('Patrón Estacional Aplicado al Forecast', fontsize=14, fontweight='bold')
plt.xlabel('Mes', fontsize=13)
plt.ylabel('Factor Estacional', fontsize=13)
plt.xticks(range(12), meses, rotation=45)
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

print("\n✓ Modelo ejecutado exitosamente")
print("✓ Gráficas generadas")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
from sklearn.linear_model import LinearRegression
import requests
from io import BytesIO
import warnings
warnings.filterwarnings('ignore')

# ============================================================
# CONFIGURACIÓN
# ============================================================
COLUMNA_DATOS = 'L'  # Cambiar a 'C', 'R', etc. según necesites
PERCENTIL_INFERIOR = 10  # Percentil inferior para winsorización
PERCENTIL_SUPERIOR = 90  # Percentil superior para winsorización
# ============================================================

# Descargar el archivo
url = "https://github.com/santiagonajera/MODELACION-Y-PRONOSTICOS-DE-LA-DEMANDA/raw/refs/heads/main/VentasHistoricas-CursoForecast-nuevo.xlsx"
response = requests.get(url)
data = BytesIO(response.content)

# Leer columna A (fechas) y la columna especificada
columnas = f'A,{COLUMNA_DATOS}'
df = pd.read_excel(data, sheet_name='FORECAST-Ventas', usecols=columnas)
df.columns = ['Mes', 'Ventas_Original']

print(f"Usando columna {COLUMNA_DATOS} para el análisis")
print(f"Total de registros: {len(df)}")

# ============================================================
# WINSORIZACIÓN DE DATOS
# ============================================================
# Calcular percentiles
percentil_inf = np.percentile(df['Ventas_Original'], PERCENTIL_INFERIOR)
percentil_sup = np.percentile(df['Ventas_Original'], PERCENTIL_SUPERIOR)

# Aplicar winsorización
df['Ventas_Winsorizada'] = df['Ventas_Original'].copy()
df.loc[df['Ventas_Winsorizada'] < percentil_inf, 'Ventas_Winsorizada'] = percentil_inf
df.loc[df['Ventas_Winsorizada'] > percentil_sup, 'Ventas_Winsorizada'] = percentil_sup

# Contar valores extremos
valores_inferiores = (df['Ventas_Original'] < percentil_inf).sum()
valores_superiores = (df['Ventas_Original'] > percentil_sup).sum()

print(f"\nWINSORIZACIÓN APLICADA:")
print("=" * 50)
print(f"Percentil inferior ({PERCENTIL_INFERIOR}%): ${percentil_inf:,.2f}")
print(f"Percentil superior ({PERCENTIL_SUPERIOR}%): ${percentil_sup:,.2f}")
print(f"Valores extremos inferiores ajustados: {valores_inferiores}")
print(f"Valores extremos superiores ajustados: {valores_superiores}")

# Verificar si hay valores cero o negativos
min_valor = df['Ventas_Winsorizada'].min()
tiene_ceros_negativos = min_valor <= 0

# Determinar el tipo de modelo a usar
if tiene_ceros_negativos:
    modelo_tipo = 'additive'
    print(f"\n⚠️ ADVERTENCIA: Se detectaron valores <= 0 (mínimo: ${min_valor:,.2f})")
    print(f"   Usando modelo ADITIVO en lugar de multiplicativo")
else:
    modelo_tipo = 'multiplicative'
    print(f"\n✓ Usando modelo MULTIPLICATIVO")

print("=" * 50)

# ============================================================
# FORECAST CON DATOS WINZORIZADOS
# ============================================================
# Descomposición con modelo apropiado
decomposition = seasonal_decompose(df['Ventas_Winsorizada'], model=modelo_tipo, period=12)

# 1. Extraer factores estacionales (últimos 12 meses)
seasonal_pattern = decomposition.seasonal.dropna()
seasonal_factors = seasonal_pattern[-12:].values

# 2. Proyectar la tendencia usando regresión lineal
trend_data = decomposition.trend.dropna()
X = np.arange(len(trend_data)).reshape(-1, 1)
y = trend_data.values

model = LinearRegression()
model.fit(X, y)

# Proyectar tendencia para los próximos 12 meses
future_X = np.arange(len(trend_data), len(trend_data) + 12).reshape(-1, 1)
future_trend = model.predict(future_X)

# 3. Calcular forecast futuro según el tipo de modelo
if modelo_tipo == 'multiplicative':
    # Modelo multiplicativo: Tendencia × Estacionalidad
    forecast_futuro = future_trend * seasonal_factors
else:
    # Modelo aditivo: Tendencia + Estacionalidad
    forecast_futuro = future_trend + seasonal_factors

# ============================================================
# GRÁFICA: Original + Winsorizada + Forecast
# ============================================================
plt.figure(figsize=(18, 7))

n_historico = len(df)
n_futuro = 12

# Serie original (en gris más tenue)
plt.plot(range(n_historico), df['Ventas_Original'],
         label='Serie Original', color='#CCCCCC', linewidth=2, marker='o',
         markersize=4, alpha=0.6)

# Serie winsorizada
plt.plot(range(n_historico), df['Ventas_Winsorizada'],
         label='Serie Winsorizada', color='#2E86AB', linewidth=2.5,
         marker='s', markersize=4)

# Forecast de 12 meses
future_indices = range(n_historico, n_historico + n_futuro)
plt.plot(future_indices, forecast_futuro,
         label='Forecast (12 meses)', color='#F18F01',
         linewidth=3, marker='D', markersize=7, linestyle='--')

# Línea divisoria entre histórico y forecast
plt.axvline(x=n_historico-0.5, color='red', linestyle=':', linewidth=2, alpha=0.5)

# Sombrear área de forecast
plt.axvspan(n_historico-0.5, n_historico + n_futuro, alpha=0.1, color='orange')

# Marcar valores extremos ajustados
extremos_inf = df[df['Ventas_Original'] < percentil_inf].index
extremos_sup = df[df['Ventas_Original'] > percentil_sup].index

if len(extremos_inf) > 0:
    plt.scatter(extremos_inf, df.loc[extremos_inf, 'Ventas_Original'],
                color='red', s=100, marker='v', zorder=5, alpha=0.7,
                label='Valores extremos inferiores')

if len(extremos_sup) > 0:
    plt.scatter(extremos_sup, df.loc[extremos_sup, 'Ventas_Original'],
                color='darkred', s=100, marker='^', zorder=5, alpha=0.7,
                label='Valores extremos superiores')

titulo_modelo = "Multiplicativo" if modelo_tipo == 'multiplicative' else "Aditivo"
plt.title(f'Serie Original vs Winsorizada vs Forecast ({titulo_modelo}) - Columna {COLUMNA_DATOS}',
          fontsize=16, fontweight='bold')
plt.xlabel('Tiempo (meses)', fontsize=13)
plt.ylabel('Ventas', fontsize=13)
plt.legend(fontsize=11, loc='best')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

# ============================================================
# RESULTADOS
# ============================================================
print(f"\nFORECAST PRÓXIMOS 12 MESES (Modelo {titulo_modelo}):")
print("=" * 50)
meses = ['Ene', 'Feb', 'Mar', 'Abr', 'May', 'Jun',
         'Jul', 'Ago', 'Sep', 'Oct', 'Nov', 'Dic']
for i, mes in enumerate(meses):
    print(f"{mes}: ${forecast_futuro[i]:,.2f}")
print("=" * 50)
print(f"Total proyectado: ${forecast_futuro.sum():,.2f}")
print(f"Promedio mensual: ${forecast_futuro.mean():,.2f}")

print("\nESTADÍSTICAS COMPARATIVAS:")
print("=" * 50)
print(f"Media Original: ${df['Ventas_Original'].mean():,.2f}")
print(f"Media Winsorizada: ${df['Ventas_Winsorizada'].mean():,.2f}")
print(f"Desv. Est. Original: ${df['Ventas_Original'].std():,.2f}")
print(f"Desv. Est. Winsorizada: ${df['Ventas_Winsorizada'].std():,.2f}")
print(f"Mínimo Winsorizado: ${df['Ventas_Winsorizada'].min():,.2f}")
print(f"Máximo Winsorizado: ${df['Ventas_Winsorizada'].max():,.2f}")
print("=" * 50)
```
