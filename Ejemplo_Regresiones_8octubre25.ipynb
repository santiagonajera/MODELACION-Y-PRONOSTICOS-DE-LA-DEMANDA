```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Cargar los datos desde el archivo Excel
df = pd.read_excel('VentasHistoricas-CursoForecast-nuevo.xlsx', sheet_name='FORECAST-Ventas', usecols='A:B')

# Asumiendo que la columna A es el tiempo (X) y la columna B es el valor (Y)
X = np.arange(len(df)).reshape(-1, 1)  # Usar índice numérico para X
Y = df.iloc[:, 1].values  # Valores de Y

# Calcular la regresión lineal
model = LinearRegression()
model.fit(X, Y)

# Obtener los coeficientes beta_0 y beta_1
beta_0 = model.intercept_
beta_1 = model.coeficiente_[0]

# Imprimir la ecuación de la regresión
print(f"Ecuación de la regresión lineal: Y = {beta_0:.2f} + {beta_1:.2f} * X")

# Graficar los datos y la línea de regresión
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
sns.regplot(x=np.arange(len(df)), y=Y, ci=None, color='darkblue', line_kws={'color': 'red'})

# Personalizar el gráfico
plt.title('Tendencia de Ventas', fontsize=16)
plt.xlabel('Tiempo (meses)', fontsize=14)
plt.ylabel('Ventas', fontsize=14)
plt.xticks(np.arange(len(df)), df.iloc[:, 0], rotation=45)
plt.tight_layout()

# Mostrar el gráfico
plt.show()
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
from sklearn.linear_model import LinearRegression
import requests
from io import BytesIO

# Descargar el archivo
url = "https://github.com/santiagonajera/MODELACION-Y-PRONOSTICOS-DE-LA-DEMANDA/raw/refs/heads/main/VentasHistoricas-CursoForecast-nuevo.xlsx"
response = requests.get(url)
data = BytesIO(response.content)
df = pd.read_excel(data, sheet_name='FORECAST-Ventas', usecols='A:B')

# Asignar nombres a las columnas
df.columns = ['Mes', 'Ventas']

# Convertir a serie de tiempo (índice numérico)
df['Mes_num'] = np.arange(len(df))
df.set_index('Mes_num', inplace=True)

# Descomposición multiplicativa
decomposition = seasonal_decompose(df['Ventas'], model='multiplicative', period=12)  # Asumiendo datos mensuales

# Crear DataFrame con los componentes
result_df = pd.DataFrame({
    'Tendencia': decomposition.trend,
    'Estacionalidad': decomposition.seasonal,
    'Residuo': decomposition.resid
})

# Mostrar el DataFrame con los componentes
print(result_df.head())

# --- Gráficos ---
sns.set(style="whitegrid", palette="muted")

# Gráfico de la tendencia
plt.figure(figsize=(12, 5))
sns.lineplot(data=df['Ventas'], label='Serie Original', color='gray', alpha=0.5)
sns.lineplot(data=decomposition.trend, label='Tendencia', color='darkblue', linewidth=2)
plt.title('Tendencia de la Serie de Ventas', fontsize=16)
plt.xlabel('Tiempo (meses)', fontsize=14)
plt.ylabel('Ventas', fontsize=14)
plt.legend()
plt.tight_layout()
plt.show()

# Gráfico de la estacionalidad
plt.figure(figsize=(12, 5))
sns.lineplot(data=decomposition.seasonal, label='Estacionalidad', color='darkred', linewidth=2)
plt.title('Estacionalidad Multiplicativa de la Serie de Ventas', fontsize=16)
plt.xlabel('Tiempo (meses)', fontsize=14)
plt.ylabel('Factor Estacional', fontsize=14)
plt.legend()
plt.tight_layout()
plt.show()
```
